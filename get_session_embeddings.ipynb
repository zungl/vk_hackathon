```python
import pandas as pd
import pickle 
import pathlib
import numpy as np
from tqdm.notebook import tqdm_notebook
```

```python
import networkx as nx
from node2vec import Node2Vec;
```

```python
base = pathlib.Path("data/graph")
emb_savepath = pathlib.Path("embeddings")

_, edgelist_path, _ = sorted(base.glob("*"))

edgelist = pd.read_csv(edgelist_path)
```

```python
edges_list = edgelist[['from', 'to']].to_numpy()[:,:2]
weights = list(map(lambda x: {x[0]:x[1]}, 
                   zip(["weight" for j in range(edgelist.shape[0])],
                       edgelist["w"].tolist())))
```

```python
G = nx.DiGraph()
G.add_edges_from(edges_list, weigth=edgelist.w.to_numpy().reshape(-1,1))
```

```python
G = nx.DiGraph()
for idx, link in tqdm_notebook(enumerate(edges_list)):
    link = list(link)
    G.add_edge(*link, 
#                weight=np.array([weights[idx]], dtype=np.float64)
               weight=float(weights[idx]['weight'])
              )
# G.add_edges_from(edges_list, weight=edgelist['w'].tolist())
```

```python
# !pip install nodevectors
```

```python
import nodevectors
ggvec_model = nodevectors.GGVec() 
embeddings = ggvec_model.fit_transform(G)
```

```python
embeddings = pd.DataFrame(embeddings, 
                          columns=[f"x_{i}" for i in range(embeddings.shape[1])])
embeddings["session_id"] = range(embeddings.shape[0])
```

```python
embeddings.set_index("session_id")
```
